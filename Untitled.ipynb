```python
import pandas as pd
import numpy as np
import itertools
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics

```

```python
df=pd.read_excel('Book2.xlsx')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x=df[["Service"]]
x
```

```python
y=df[["Region","Project","Approach","Results"]]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)
```

```python
Tfidf_Vectorizer=TfidfVectorizer()
Tfidf_train=Tfidf_Vectorizer.fit_transform(X_train)
print(Tfidf_train)
Tfidf_test=Tfidf_Vectorizer.transform(X_test)
```

```python
Tfidf_Vectorizer = TfidfVectorizer()
Tfidf_train = Tfidf_Vectorizer.fit_transform(X_train["Service"])
Tfidf_test = Tfidf_Vectorizer.transform(X_test["Service"])
```

```python
clf = MultinomialNB()
```

```python
from sklearn.multioutput import MultiOutputClassifier
multi_target_clf = MultiOutputClassifier(clf)
multi_target_clf.fit(Tfidf_train, y_train)
pred = multi_target_clf.predict(Tfidf_test)
```

```python
from sklearn.metrics import hamming_loss


hamming_losses = [hamming_loss(y_test[col], pred[:, idx]) for idx, col in enumerate(y_test.columns)]


for idx, col in enumerate(y_test.columns):
    print(f'Hamming Loss for {col}: {hamming_losses[idx]:.3f}')
```

```python
average_hamming_loss = sum(hamming_losses) / len(hamming_losses)
print(f'Average Hamming Loss: {average_hamming_loss:.3f}')
```

```python
from sklearn.metrics import accuracy_score

accuracy_scores = [accuracy_score(y_test[col], pred[:, idx]) for idx, col in enumerate(y_test.columns)]


for idx, col in enumerate(y_test.columns):
    print(f'Accuracy Score for {col}: {accuracy_scores[idx]:.3f}')
```

```python
from sklearn.metrics import classification_report


for col in y_test.columns:
    report = classification_report(y_test[col], pred[:, y_test.columns.get_loc(col)])
    print(report)
```
